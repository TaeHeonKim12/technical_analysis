```python
import pandas as pd
from datetime import datetime
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/01_CME_Mini_NQ.csv")
df1.info()
df1
```

```python
df1['time'] = pd.to_datetime(df1['time'],unit='s')
```

```python
from scipy.signal import argrelextrema
import numpy as np
```

```python
s=300
e=500
i=4
dfpl = df1[s:e]

max_idx = (argrelextrema(dfpl['high'].values, np.greater_equal, order=i)[0])
min_idx = (argrelextrema(dfpl['low'].values, np.less_equal, order=i)[0])

dfpl = dfpl.reset_index()

print(len(max_idx), len(min_idx))
```

```python
HorL = []
for i in range(len(max_idx)):
  if i==0:
    HorL.append([max_idx[i], dfpl['high'][max_idx[i]], 'start'])
  elif dfpl['high'][max_idx[i]] > dfpl['high'][max_idx[i-1]]:
    HorL.append([max_idx[i], dfpl['high'][max_idx[i]], 'HH'])
  elif dfpl['high'][max_idx[i]] < dfpl['high'][max_idx[i-1]]:
    HorL.append([max_idx[i] , dfpl['high'][max_idx[i]],'HL'])
  else:
    HorL.append([max_idx[i], dfpl['high'][max_idx[i]], 0])

LorH = []
for i in range(len(min_idx)):
  if i==0:
    LorH.append([min_idx[i], dfpl['low'][min_idx[i]], 'start'])
  elif dfpl['low'][min_idx[i]] > dfpl['low'][min_idx[i-1]]:
    LorH.append([min_idx[i], dfpl['low'][min_idx[i]], 'LH'])
  elif dfpl['low'][min_idx[i]] < dfpl['low'][min_idx[i-1]]:
    LorH.append([min_idx[i] , dfpl['low'][min_idx[i]],'LL'])
  else:
    LorH.append([min_idx[i], dfpl['low'][min_idx[i]], 0])


```

```python
HorL = pd.DataFrame(HorL)
HorL
```

```python
LorH = pd.DataFrame(LorH)
LorH
```

```python
ee = pd.concat([HorL,LorH], axis=0)
ee
```

```python
ee.columns= ['i','price','trend']
ee = ee.sort_values('i')
ee = ee.reset_index(drop=True)
ee
```

```python
ee1 =ee
t = []
for i in range(1, len(ee1)):
  if (ee1['trend'][i-1] == 'LL' and ee1['trend'][i] == 'LH') or (ee1['trend'][i-1] == 'LH' and ee1['trend'][i] == 'LL') or (ee1['trend'][i-1] == 'LL' and ee1['trend'][i] == 'LL') or (ee1['trend'][i-1] == 'LH' and ee1['trend'][i] == 'LH'):
    a= ee1['i'][i-1]
    b = ee1['i'][i]
    highvalue = max(dfpl['high'][a+1:b])
    df_highvalue = dfpl[a+1:b]
    c = df_highvalue[df_highvalue['high']== highvalue].index[0]
    t.append([c, highvalue, 'H'])
  elif (ee1['trend'][i-1] == 'HL' and ee1['trend'][i] == 'HH') or (ee1['trend'][i-1] == 'HH' and ee1['trend'][i] == 'HL') or (ee1['trend'][i-1] == 'HH' and ee1['trend'][i] == 'HH') or (ee1['trend'][i-1] == 'HL' and ee1['trend'][i] == 'HL'):
    a= ee1['i'][i-1]
    b = ee1['i'][i]
    lowvalue = min(dfpl['low'][a+1:b])
    df_lowvalue = dfpl[a+1:b]
    c = df_lowvalue[df_lowvalue['low']== lowvalue].index[0]
    for l in range(len(min_idx)):
      if min_idx[l] < c:
        min_idx1 = min_idx[l]
    if lowvalue > dfpl['low'][min_idx1]:
      z = 'LH'
      t.append([c, lowvalue, z])
    elif lowvalue < dfpl['low'][min_idx1]:
      z = 'LL'
      t.append([c, lowvalue, z])
    else:
      z = 0
      t.append([c, lowvalue, z])
```

```python
t = pd.DataFrame(t)
t.columns= ['i','price','trend']
tt = pd.concat([ee1,t], axis=0)
tt = tt.sort_values('i')
tt = tt.reset_index(drop=True)
tt
```

```python
changetrend = [0] * len(tt)

for i in range(2, len(tt)):
  if  tt['trend'][i-1] =='HL' and tt['trend'][i] =='LL':
    changetrend[i-1] = 'down'
  elif tt['trend'][i-1] =='LL' and tt['trend'][i] =='HL':
    changetrend[i-1] = 'down'
  elif tt['trend'][i-1] =='LH' and tt['trend'][i] =='HH':
    changetrend[i-1] = 'up'
  elif tt['trend'][i-1] =='HH' and tt['trend'][i] =='LH':
    changetrend[i-1] = 'up'
  else:
    changetrend[i-1]=0
```

```python
tt['changetrend'] = changetrend
tt
```
